```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

df = pd.read_excel (r'Fit_17_LowChisq.xlsx', sheet_name='Relative_Deviation')

# print (df)

print(df.keys())
```

```python
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 

ax = df.plot.scatter(x="t1 (s)", y="RepA(1-70)-(TitinI27)1-Cys-AF555", yerr=df["stdev1"], linewidths = 5, color="Red", label="RepA(1-70)-(TitinI27)1-Cys-AF555", figsize = (30, 10))
df.plot.scatter(x="t2 (s)", y="RepA(1-70)-(TitinI27)2-Cys-AF555", yerr=df["stdev2"], linewidths = 5, color="Blue", label="RepA(1-70)-(TitinI27)2-Cys-AF555", ax=ax)
df.plot.scatter(x="t3 (s)", y="RepA(1-70)-(TitinI27)3-Cys-AF555", yerr=df["stdev3"], linewidths = 5, color="DarkGreen", label="RepA(1-70)-(TitinI27)3-Cys-AF555", ax=ax);

df.plot( x = "t1_sim (s)" , y = "Simulated data 1", linewidth=5, color="Red", ax = ax )
df.plot( x = "t2_sim (s)" , y = "Simulated data 2", linewidth=5, color="Blue", ax = ax )
df.plot( x = "t3_sim (s)" , y = "Simulated data 3", linewidth=5, color="DarkGreen", ax = ax )

plt.xlabel('t (s)', fontsize=24)
plt.ylabel('Relative Deviation in Fluorescence (A.U.)', fontsize=24)
plt.legend(prop={'size': 24})

plt.xlim([0, 805])
plt.savefig('kendkc_750uMATP_Global_n_Fit_17_LowChisq.png', dpi=300, bbox_inches='tight')
plt.savefig('kendkc_750uMATP_Global_n_Fit_17_LowChisq.eps', dpi=300, bbox_inches='tight')
plt.show()
```
